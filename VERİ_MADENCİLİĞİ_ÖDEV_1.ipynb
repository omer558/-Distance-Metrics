# **Euclidean Distance**


Öklid mesafesi iki gerçek değerli vektör arasındaki mesafeyi hesaplar.Sayısal değerlere sahip iki veri satırı arasındaki mesafeyi hesaplarken  **Öklid mesafesini** kullanırız.
Çoğu makine öğrenimi algoritması, gözlemler arasındaki benzerliği ölçmek için bu mesafe metriğini kullanır. Diyelim ki aşağıda gösterildiği gibi iki noktamız var:

Öklid mesafesi, iki vektör arasındaki kare farkların toplamının karekökü olarak hesaplanır.
Mesafe hesaplaması binlerce veya milyonlarca kez yapılacaksa, hesaplamayı hızlandırmak amacıyla karekök işleminin kaldırılması uygundur. Elde edilen sonuçlar, bu değişiklikten sonra aynı oranlara sahip olacak ve en benzer örnekleri bulmak için bir makine öğrenimi algoritması içinde etkili bir şekilde kullanılabilir.

```python
from math import sqrt
 
def Euclidean_distance(a, b):
	return sqrt(sum((x1-x2)**2 for x1, x2 in zip(a,b)))
 
data1 = [0,57,0,57]
data2 = [57,0,57,57]

uzaklık = Euclidean_distance(data1,data2)
print('mesafe:',uzaklık)
```

Bu kodda karekök ifadesini math kütüphanesinden import edip yeni bir Euclidian_distance fonksiyonu oluşturduk.Fonksiyon her çağrıldığında sonucu bize girdiğimiz vektör listelerinin elemanlarını sırasıyla zip komutuyla eşleyip,elemanların farklarının karelerini aldıktan sonra toplamlarının karekökünü döndürecektir.

# **Manhattan Distance**


Manhattan mesafesi, iki nokta arasındaki mesafenin Kartezyen koordinatlarının mutlak farklarının toplamı olarak hesaplandığı bir ölçüdür . Basit bir deyişle, x koordinatları ve y koordinatları arasındaki farkın toplamıdır.

Aşağıdaki kodda bu formülü kullanarak Manhattan mesafesini hesaplayacağız:

```python
from math import sqrt
 
def manhattan_distance(a, b):
	return sum(abs(float(x2)-float(x1))for x1,x2 in zip(a,b))
 
data1 = [28,12,42,52]
data2 = [88,54,75,65]

uzaklık = manhattan_distance(data1,data2)
print('mesafe:',uzaklık)
```

Bu kodda karekök ifadesini math kütüphanesinden import edip yeni bir Manhattan_distance fonksiyonu oluşturduk.Fonksiyon her çağrıldığında sonucu bize girdiğimiz listelerin elemanlarını sırasıyla zip komutuyla eşleyip,elemanların farklarının abs komutu ile mutlak değerini aldıktan sonra toplamlarını döndürecektir.

# **Minkowski Distance**

Minkowski Mesafesi, Öklid ve Manhattan Mesafesinin genelleştirilmiş şeklidir.Farklı mesafe ölçülerinin hesaplanmasına izin veren "p" adı verilen bir parametre ekler.

Minkowski mesafe ölçüsü şu şekilde hesaplanır:

image.png

Formülde görüldüğü üzere p değerine 1 verilir ise Manhattan mesafesini ,2 verilir ise Öklid mesafesinin formülüyle aynı olur.Minkowski mesafesi p yerine farklı değerler verilmesiyle farklı uzaklıklarıda hesaplar.

Minkowski Mesafesini aşağıdaki kod ile hesaplayabiliriz:

```python
from math import sqrt
 
def minkowski_distance(a, b, p):
	return sum(abs(x1-x2)**p for x1, x2 in zip(a,b))**(1/p)
 
data1 = [1, 2, 3, 4]
data2 = [4, 6, 8, 2]
# p=1 için hesaplayalım
uzaklık = minkowski_distance(data1,data2,1)
print('mesafe:',uzaklık)
# p=1 için hesaplayalım
uzaklık = minkowski_distance(data1,data2,2)
print('mesafe:',uzaklık)
```

Yukarıdaki kodda karekök ifadesini math kütüphanesinden import edip yeni bir Minkowski_distance fonksiyonu oluşturduk.Fonksiyon her çağrıldığında sonucu bize girdiğimiz listelerin elemanlarını sırasıyla zip komutuyla eşleyip,elemanların farklarının abs komutu ile mutlak değerini aldıktan sonra eğer 'p' komutu 1 alındığında elemanların farklarının abs komutu ile mutlak değerini aldıktan sonra toplamlarını döndürecektir.'p' 2 alındığında elemanların farklarının mutlak değerinin karesi aldıktan sonra toplamlarının karekökünü döndürecektir.

# **Hamming Distance**

Hamming mesafesi, ikili veri dizisini karşılaştırmak için bir ölçüdür. Eşit uzunluktaki iki ikili dizgiyi karşılaştırırken, Hamming mesafesi, iki bitin farklı olduğu bit konumlarının sayısıdır.Adını Amerikalı matematikçi Richard Hamming'den almıştır.

image.png

```python
def hamming_distance(string1, string2):
  distance = 0 
  for n in range(len(string1)):
     if string1[n] != string2[n]:
       distance += 1
  return distance
```

```python
hamming_distance('1011','0101')
```

```python
hamming_distance('euclidean','manhattan')
```

ilk olarak hamming_distance fonksiyonunu tanımlıyoruz. 'Distance' değerine 0 atadıkdan sonra for döngüsüyle fonksiyon string indexlerini n kutusuna yolluyoruz. Eğer string indexlerindeki  n değerleri birbirinden farklıysa 'distance' değerini 1 arttırıyoruz. En son adımda ise fonksiyon çağrıldığında 'distance' değerini return ediyoruz. 

# **Cosine Similarity**

Kosinüs benzerliği, boyutlarına bakılmaksızın belgelerin ne kadar benzer olduğunu ölçmek için kullanılan bir ölçüdür.Döküman analizinde kullanılır her kelime vektör olarak tanımlanır. Matematiksel olarak, çok boyutlu bir uzayda yansıtılan iki vektör arasındaki açının kosinüsünü ölçer. Kosinüs benzerliği avantajlıdır, çünkü iki benzer belge Öklid mesafesinden çok uzak olsa bile, yine de birbirlerine daha yakın yönlendirilmiş olma ihtimalleri vardır. Açı ne kadar küçükse kosinüs benzerliği o kadar yüksek olur.

Bir benzerlik ölçütü olarak, kosinüs benzerliği genel kelime sayısından nasıl farklıdır?

Her boyutun belgedeki bir kelimeye karşılık geldiği çok boyutlu bir alana çizildiğinde, kosinüs benzerliği belgelerin büyüklüğünü değil yönünü (açısını) yakalar. Büyüklüğü istiyorsanız, bunun yerine Öklid mesafesini hesaplayın.

Cosine Similarity fonksiyonunu aşağıdaki kodda tanımladık:

```python
from math import sqrt

def norm(x):
  
  return round(sqrt(sum([a*a for a in x])),2)

def cosine_similarity(x,y):

  ic_carpım = sum(a*b for a,b in zip(x,y))
  norm_carpımları = norm(x) * norm(y)
  return round(ic_carpım/float(norm_carpımları),2)


print(cosine_similarity([0,57,0,57],[57,0,57,57]))

 
```

Kodumuzda **Cosine_similarity** fonksiyonunu tanımlamak için öncelikle vektör normunu hesaplayan fonksiyon tanımladık. Daha sonrasında vektörlerin iç çarpım değerini tanımladık ve bir alt satırında vekör normlarının çarpımlarını tanımladık.Sonuç olarak bu fonksiyon bize vektörlerin iç çarpımının , boylarına çarpımının oranını döndürecek.

# **Correlation**

İki veya daha fazla bağımsız değişken arasındaki ilişkinin varlığı, bu ilişkinin yönü ve şiddeti korelasyon analizi ve sonucunda elde edilen korelasyon katsayısı ile belirlenir. Formülü aşağıdaki gibidir.



image.png

corr(x,y) = covariance(x,y) / s.deviation(x)+s.deviation(y)

```python
from math import sqrt

def ortalama(x):
  
  return round(sum([ a / len(x) for a in x]),2)
  
def standart_sapma(x):

  return round(sqrt (sum([(a - ortalama(x))**2 for a in x])),2)

def cov(x,y):
  n= len(x)
  return (sum((a - ortalama(x))*(b - ortalama(y)) for a,b in zip(x,y)))/n

def corr(x,y):

  return round((cov(x,y)/(standart_sapma(x)*standart_sapma(y))),3)


print(corr([1,2,3,4],[4,8,12,16]))
```
